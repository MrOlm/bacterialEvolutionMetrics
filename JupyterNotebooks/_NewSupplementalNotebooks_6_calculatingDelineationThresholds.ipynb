```python
def calculate_thresholds_3(Odb, m, values, within, without, col='ani', MIN_VAL=0, gene='test', gl='test'):
    '''
    Calculate thresholds based on distance matrix.
    
    Version 2 ignores 'unk'
    
    Args:
        Odb: comparison matrix
        cols: columns of the matrix to run on with ID values
        metrics: columns of the matrix that are either True/False
        values: threshold values to use
    ''' 
    within = len(Odb[(Odb[m] == True)])
    without = len(Odb[(Odb[m] == False)])
    
    if without == 0:
        return
    
    for v in values:
        db = Odb[Odb[col] >= v]
        
        if len(db) == 0:
            precision = np.nan
        else: 
            precision = len(db[db[m] == True]) / len(db)

        recall = len(db[db[m] == True]) / within
        fpr = len(db[db[m] == False]) / without

        table['genomeList'].append(gl)
        table['gene'].append(gene)
        table['threshold'].append(v)
        table['method'].append(col)
        table['metric'].append(m)
        table['precision'].append(precision)
        table['recall'].append(recall)
        table['fpr'].append(fpr)
        table['tpr'].append(recall)

# TEST
# table = defaultdict(list)
# values = list(np.arange(0.9,1.001,.001))

# within_comps = cdb[cdb['list'] == 'infantList']['within'].tolist()[0]
# without_comps = cdb[cdb['list'] == 'infantList']['without'].tolist()[0]

# calculate_thresholds_3(tdb, 'same_cluster', values, within_comps, without_comps, col='ani')

# pd.DataFrame(table)
```
